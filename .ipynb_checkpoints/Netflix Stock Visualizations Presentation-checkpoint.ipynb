```python
#Import files, libraries, and rename columns
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

netflix_stocks = pd.read_csv('NFLX.csv')
dowjones_stocks = pd.read_csv('DJI.csv')
netflix_stocks_quarterly = pd.read_csv('NFLX_daily_by_quarter.csv')

netflix_stocks.rename(columns={'Adj Close':'Price'}, inplace=True)
dowjones_stocks.rename(columns={'Adj Close':'Price'}, inplace=True)
netflix_stocks_quarterly.rename(columns={'Adj Close':'Price'}, inplace=True)
```

```python
# Creates violinplot showing Distribution of 2017 Netflix Stock Prices by Quarter
plt.figure(figsize=(7,5))
ax = sns.violinplot(x='Quarter', y='Price', data=netflix_stocks_quarterly)
ax.set_title('Distribution of 2017 Netflix Stock Prices by Quarter')
ax.set_ylabel('Closing Stock Price')
ax.set_xlabel('Business Quarters in 2017')
plt.show()
```

In 2017, median Netflix stock prices showed a steady increase across quarters.  Prices in Quarter 1 began around `$`120 and ended the year close to `$`200.  

```python
# Create a scatterplot to compare the Actual and Estimated Earnings each Quarter of 2017
x_positions = [1, 2, 3, 4]
chart_labels = ["1Q2017","2Q2017","3Q2017","4Q2017"]
earnings_actual =[.4, .15,.29,.41]
earnings_estimate = [.37,.15,.32,.41 ]
```

```python
plt.figure(figsize=(7,5))
plt.scatter(x=x_positions, y=earnings_actual, color='red', alpha=0.5)
plt.scatter(x=x_positions, y=earnings_estimate, color='blue', alpha=0.5)
plt.legend(['Actual', 'Estimate'])
plt.xticks(x_positions, chart_labels)
plt.title('Earnings Per Share in Cents')
plt.show()
```

The actual earnings of Netflix were fairly consistent with the estimated earnings across the year. Quarter 2 and Quarter 4 showed exact matches, while Quarter 1 showed a few cents earned over the estimate, and Quarter 3 showed a few cents less earned when compared to the estimate.

```python
# Create a bar graph to compare the amount of Revenue and Earnings for Netflix in 2017
# The metrics below are in billions of dollars
revenue_by_quarter = [2.79, 2.98,3.29,3.7]
earnings_by_quarter = [.0656,.12959,.18552,.29012]
quarter_labels = ["2Q2017","3Q2017","4Q2017", "1Q2018"]

# Revenue
n = 1  # This is our first dataset (out of 2)
t = 2 # Number of dataset
d = 4 # Number of sets of bars
w = 0.8 # Width of each bar
bars1_x = [t*element + w*n for element
             in range(d)]
# Earnings
n = 2  # This is our second dataset (out of 2)
t = 2 # Number of dataset
d = 4 # Number of sets of bars
w = 0.8 # Width of each bar
bars2_x = [t*element + w*n for element
             in range(d)]

labels = ["Revenue", "Earnings"]
```

```python
plt.figure(figsize=(7, 4))
plt.bar(bars1_x, revenue_by_quarter)
plt.bar(bars2_x, earnings_by_quarter)
middle_x = [ (a + b) / 2.0 for a, b in zip(bars1_x, bars2_x)]
plt.legend(labels)
plt.title('Netflix Revenue and Earnings in Billions in 2017')
plt.xticks(middle_x, quarter_labels)
plt.show()
```

Netflix demonstrated an increasing trend for both Revenue and Earnings across the 2017 year.

```python
# Netflix plot
plt.figure(figsize=(7, 4))
ax1 = plt.subplot(1, 1, 1)
plt.plot(netflix_stocks['Date'], netflix_stocks['Price'])
ax1.set_title('Netflix')
ax1.set_xlabel('Date')
ax1.set_ylabel('Stock Price')
plt.xticks(rotation=30)
plt.show()
```

```python
# Dow Jones Plot
plt.figure(figsize=(7, 4))
ax2 = plt.subplot(1, 1, 1)
plt.plot(dowjones_stocks['Date'], dowjones_stocks['Price'])
ax2.set_title('Dow Jones')
ax2.set_xlabel('Date')
ax2.set_ylabel('Stock Price')
plt.xticks(rotation=30)
plt.show()
```

The price of Netflix stock was a bit more volatile than that of the Dow Jones Average, but demonstrated an overall increasing trend across the year.  From April to June, Netflix experienced a drop in value while the Dow continued to improve slightly.  June showed a big jump in stock prices for Netflix, and again from August through October, while the Dow showed a steady increase through the year.  For approximately each `$`1000 increase in the Dow Jones Average, Netflix experienced about a `$`10 increase in stock price.
